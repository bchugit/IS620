```python
df_districts = pandas.read_csv(dataDir + '/top_ip_logins.csv')
df_districts
```

```python
# all words

words = list()
words_list = df_districts.dropna().name.tolist()

for ww in words_list:
    words += ww.split()
```

```python
from nltk.corpus import movie_reviews
#all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
all_words = nltk.FreqDist(w.lower() for w in words)
word_features = list(all_words)[:2000] # [_document-classify-all-words]

def document_features(document): # [_document-classify-extractor]
    document_words = set(document) # [_document-classify-set]
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features

```

```python
# create featureset for all rows
featuresets = [(document_features(row['name'].lower().split()), row['test']) for (ii,row) in df_districts.dropna().iterrows()]
```

```python
train_set, test_set = featuresets[50:], featuresets[:50]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

```python
# pick a single row to classify
n = 4

print df_districts.name[n]
classifier.classify(featuresets[n][0]), df_districts['test'][n]
```

```python
# classify all of them at once
df_districts['predict'] = None
df_districts['predict'][:50] = classifier.classify_many([ff for ff,cc in test_set])
```
